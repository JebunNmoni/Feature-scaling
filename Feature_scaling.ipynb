### Read data from income.csv and scale the relevant features using standardization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data_set = pd.read_csv('income.csv')
data_set
```

```python
x = data_set[['Age', 'Income']]
x
```

```python
sns.distplot(x['Age'])
```

```python
x_temp = x.copy()
x_temp['Age'] = (x_temp['Age'] - x_temp['Age'].min()) / (x_temp['Age'].max() - x_temp['Age'].min())
x_temp
```

```python
sns.distplot(x_temp['Age'])
```

```python
x_temp['Income'] = (x_temp['Income'] - x_temp['Income'].min()) / (x_temp['Income'].max() - x_temp['Income'].min())
x_temp
```

```python
#import sklearn

from sklearn.preprocessing import MinMaxScaler

MMS = MinMaxScaler()

x_after_min_max_scaler = MMS.fit_transform(x)

x_after_min_max_scaler
```

```python
x_temp = x.copy()
x_temp['Age'] = (x_temp['Age'] - x_temp['Age'].mean()) / x_temp['Age'].std()
x_temp
```

```python
sns.distplot(x_temp['Age'])
```

```python
from sklearn.preprocessing import StandardScaler as SC

std = SC()

x_after_Standardisation = std.fit_transform(x)

x_after_Standardisation
```

```python
df_temp = x.copy()
df_temp['Age'] = df_temp['Age'] / df_temp['Age'].abs().max()
df_temp
```

```python
sns.distplot(df_temp['Age'])
```

```python
from sklearn.preprocessing import MaxAbsScaler

maxabs = MaxAbsScaler()

x_after_maxabs = maxabs.fit_transform(x)

x_after_maxabs
```
